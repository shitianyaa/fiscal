# 统计语言模型




先假设我们只有 5 句简单的 “训练句子”（就像小朋友学说话时听的简单话），作为模型的 “知识储备”：

- 小猫吃鱼
- 小狗跑
- 小猫睡觉
- 小狗吃骨头
- 小猫玩球

统计 “单个词出现的次数”（先摸清 “谁常出现”）
- 小猫：3 次（句子 1、3、5）
- 小狗：2 次（句子 2、4）
- 吃：2 次（句子 1、4）
- 鱼：1 次（句子 1）
- 跑：1 次（句子 2）
- 睡觉：1 次（句子 3）
- 骨头：1 次（句子 4）
- 玩球：1 次（句子 5）

“如果前一个词是‘小猫’，下一个词是‘吃鱼’的可能性有多大？”
- 前一个词是 “小猫” 时，后面接的词有 3 种（吃鱼、睡觉、玩球），各 1 次。
所以：“小猫→吃鱼” 的概率 = 1÷3 ≈ 33%；“小猫→睡觉”≈33%；“小猫→玩球”≈33%。
