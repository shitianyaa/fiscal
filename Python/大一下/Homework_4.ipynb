```python
for n in range(6, 1000, 2):
    for i in range(2, n):
        n1 = i
        n2 = n - i
        prime_n1 = True
        prime_n2 = True
        for j in range(2, n1):
            if n1 % j == 0:
                prime_n1 = False
                break
        for j in range(2,n2):
            if n2 % j == 0:
                prime_n2 = False
                break
        if prime_n1 and prime_n2:
            found = True
            break
    if not found:
        print(f"{n} 无法分解为两个质数之和")
        break
else:
    print("对于大于等于6且小于1000的所有偶数，都能分解成两个质数之和")
    
```
