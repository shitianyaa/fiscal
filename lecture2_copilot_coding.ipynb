# 虚拟环境

可以把 Python 的虚拟环境想象成 **“每个项目专属的‘隔离小房间’”**。
可以把 Python 的虚拟环境想象成 **“每个项目专属的‘隔离小房间’”**。

假设你是个 “手作达人”，同时在做三个项目：

- 项目 A：需要用胶水 A、剪刀 A、卡纸 A（对应特定版本的 Python 库，比如numpy==1.19.0）
- 项目 B：需要用胶水 B、剪刀 B、卡纸 B（对应numpy==1.24.0，和 A 不兼容）
- 项目 C：完全不用胶水，只用特殊剪刀 C（可能不需要numpy，但需要其他库）

如果没有虚拟环境，就像把所有工具堆在一个大桌子上：

- 做 A 时可能不小心用了胶水 B，导致作品坏了（库版本冲突报错）；
- 做完 B 后想清理，可能误删了 A 需要的剪刀 A（卸载依赖时影响其他项目）；
项目多了，根本记不清哪个工具属于哪个项目（依赖管理混乱）。

**大家查看并激活自己安装Anaconda自动安装的base环境**

- 方法1：打开 Anaconda 自带的终端（Windows 用Anaconda Prompt），如果命令行最前面显示 (base)，说明当前正处于base环境中。如果没有显示(base)，可以先手动激活它：conda activate base
- 方法2：在终端输入：conda env list，查看所有环境，可以激活base

## 课后学习内容1：学习使用git和github

git和github
- Git负责代码的版本管理，你可以在本地使用Git来跟踪代码的变化
- GitHub提供了一个云端空间，让使用Git管理的代码可以被上传、共享和协作
- 开发者通常在本地使用Git进行开发，然后将代码推送到GitHub上与他人共享

### 将fiscal项目上传到GitHub的步骤如下：

#### 1. 在GitHub上创建新仓库
1. 登录您的GitHub账户
2. 点击右上角的"+"号，选择"New repository"
3. 仓库名称输入"fiscal"（或您想要的其他名称）
4. 可以选择添加描述
5. 选择公开（Public）或私有（Private）
6. 点击"Create repository"

#### 2. 在本地初始化Git并推送代码
打开终端（Windows上可以使用Git Bash），执行以下命令：
```
git init
git add .
git commit -m "Initial commit"
git branch -M main  # This renames the default branch to main (optional)
git remote add origin https://github.com/haichaozheng/fiscal.git
git push -u origin main
```

#### 3. 验证上传结果
刷新GitHub页面，您应该能看到项目中的所有文件：
- `fiscal_app.py` - 主应用程序文件
- `requirements.txt` - Python依赖包列表
- `README.md` - 项目说明文件

这样您的fiscal项目就成功上传到GitHub了。以后每次修改后，只需要执行以下命令就可以更新GitHub上的代码：

```bash
git add .
git commit -m "描述您的修改"
git push origin main
```
